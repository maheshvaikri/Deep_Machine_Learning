## Algorithm Basics with Big -O Notation - Simple Explanation

#### Simple Function adding 0 to n in two different function definitions

```python
def sum1(n):
    final_sum = 0
    for i in range(n+1):
        final_sum += i
    return final_sum
```

```python
sum1(10)
```

```python
def sum2(n):
    return (n*(n+1))/2
```

```python
sum2(10)
```

### Lets time the functions to determine the best algorithm

```python
%timeit sum1(100)
```

```python
%timeit sum2(100)
```

## MatplotLib analysis of Big-O notations

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

# Setup runtime comparison
n = np.linspace(1, 10, 1000)
labels = ['Constant','Logrithmic','Linear','Log Linear','Quadriatic','Cubic','Exponential']

big_o = [np.ones(n.shape), np.log(n), n, n*np.log(n), n**2, n**3, 2**n]

# Plot setup

plt.figure(figsize=(12,10))
plt.ylim(0,50)
plt.xlim(1,10)

for i in range (len(big_o)):
    plt.plot(n, big_o[i], label = labels[i])

plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('n')
```
